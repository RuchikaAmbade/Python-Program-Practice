**Star Pattern Program In Python**

**1. Reverse Pyramid**

```python
n=5
for i in range(1,n+1):
    for j in range(1,i+1):
        print(' ',end='')
    for j in range(i,n+1):
        print('*',end=' ')
    print()
    
```

**2. Butterfly**

```python
n=5
for i in range(1,n+1):
    for j in range(1,i+1):
        if i==5 and j==5:
            break
        print('*',end=' ')
    for j in range(i,n):
        print(' ',end=' ')
    for j in range(i+1,n):
        print(' ',end=' ')
    for j in range(1,i+1):
        print('*',end=' ')
    print()

for i in range(2,n+1):
    for j in range(i,n+1):
        print('*',end=' ')
    for j in range(1,i):
        print(' ',end=' ')
    for j in range(2,i):
        print(' ',end=' ')
    for j in range(i,n+1):
        print('*',end=' ')
    print()
```

**3.Diamond**

```python
n=5
for i in range(1,n+1):
    for j in range(i,n+1):
        print(' ',end='')
    for j in range(1,i+1):
        print('*',end=' ')
    print()
for i in range(2,n+1):
    for j in range(1,i+1):
        print(' ',end='')
    for j in range(i,n+1):
        print('*',end=' ')
    print()
```

**4.Sandglass**

```python
n=5
for i in range(1,n+1):
    for j in range(1,i+1):
        print(' ',end='')
    for j in range(i,n+1):
        print('*',end=' ')
    print()
for i in range(2,n+1):
    for j in range(i,n+1):
        print(' ',end='')
    for j in range(1,i+1):
        print('*',end=' ')
    print()
```
