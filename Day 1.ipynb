**Star Pattern Program In Python**

**1. Pyramid**

```python
n = 5
for i in range(1,n+1):
    for j in range(i,n+1):
        print(' ',end='')
    for j in range(1,i+1):
        print('*',end=' ')
    print()
```

**2. Right Triangle**

```python
n = 5
for i in range(1,n+1):
    for j in range(1,i+1):
        print('*',end = ' ')
    print()
```

**3. Left Triangle**

```python
n=5
for i in range(1,n+1):
    for j in range(i,n+1):
        print(' ',end=' ')
    for j in range(1,i+1):
        print('*',end=' ')
    print()
```

**4. Right Downward Triangle**

```python
n = 5
for i in range(1,n+1):
    for j in range(i,n+1):
        print('*',end=' ')
    print()
```

**5. Left Downward Triangle**

```python
n = 5
for i in range(1,n+1):
    for j in range(1,i+1):
        print(' ',end=' ')
    for j in range(i,n+1):
        print('*',end = ' ')
    print()
```

**6. Double Hill**

```python
n = 5
for i in range(1,n+1):
    for j in range(i,n+1):
        print(' ',end = '')
    for j in range(1,i+1):
        print('*',end = ' ')
    for j in range(i+1,n+1):
        print(' ',end = ' ')
    for j in range(1,i+1):
        print('*',end=' ')
    print()
```
