**Print 'G' using '*'**

```python
for row in range(7):
    for col in range(6):
        if col==0 or (col==4 and (row!=1 and row!=2)) or ((row==0 or row==6) and (col>0 and col<4)) or (row==3 and col>2):
            print('*',end=' ')
        else:
            print(end='  ')
    print()
```

**Print 'H' using '*'**

```python
for row in range(7):
    for col in range(5):
        if col==0 or col==4 or (row==3 and (col>0 and col<4)):
            print('*',end=' ')
        else:
            print(end='  ')
    print()
```

**Print 'I' using '*'**

```python
for row in range(7):
    for col in range(5):
        if col==2 or (row==0 or row==6 and (col!=2)):
            print('*',end=' ')
        else:
            print(end='  ')
    print()
```

**Print 'J' using '*'**

```python
for row in range(7):
    for col in range(5):
        if col==2 or row==0 or (row==6 and col<3):
            print('*',end=' ')
        else:
            print(end='  ')
    print()   
```

**Print 'K' using '*'**

```python
i=0
j=4
for row in range(7):
    for col in range(5):
        if col==0 or row==col+2 :
            print('*',end=' ')
        elif (row==i and col==j):
            print('*',end=' ')
            i=i+1
            j=j-1
        else:
            print(end='  ')
    print()
```

**Print 'L' using '*'**

```python
for row in range(7):
    for col in range(5):
        if col==0 or row==6:
            print('*',end=' ')
        else:
            print(end='  ')
    print()
```
